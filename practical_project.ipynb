```python
import pandas as pd
import io

filename = list(uploaded.keys())[0]

df = pd.read_csv(io.BytesIO(uploaded[filename]))

df.head()
```

```python
import numpy as np

# Separate numerical and categorical features
numerical_features = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = df.select_dtypes(include=['object']).columns.tolist()

print("Numerical Features:", numerical_features)
print("Categorical Features:", categorical_features)
```

```python
# Data preparation (getting rid of or transforming incomplete values)
print(df.isnull().sum())
df[numerical_features] = df[numerical_features].fillna(df[numerical_features].median())
df[categorical_features] = df[categorical_features].fillna(df[categorical_features].mode().iloc[0])
```

```python
# Data preparation (getting rid of or transforming incomplete values)
df = df.drop_duplicates()
from scipy.stats import zscore

z_scores = np.abs(zscore(df[numerical_features]))
df_clean = df[(z_scores < 3).all(axis=1)]

print(f"Rows before cleaning: {df.shape[0]}, Rows after cleaning: {df_clean.shape[0]}")
```

```python
from sklearn.preprocessing import LabelEncoder
df_encoded = df.copy()
label_encoders = {}

for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])  # Convert to numeric
    label_encoders[col] = le


correlation_matrix = df_encoded.corr()

```

**Limitations of the Dataset**

While our dataset provides a strong foundation for analysis, some challenges remain:

Potential Missing Information: Some categorical variables may lack sufficient representation, leading to biased results.

Presence of Outliers: Extreme values in property prices may skew model performance.

Lack of External Factors: The dataset does not account for external market conditions like economic trends or interest rates, which significantly impact house prices.

**Machine Learning Algorithms for Regression**

To address the problem, we will experiment with the following regression models:


*   Linear Regression: A simple and interpretable model for understanding relationships between features.
*   Polynomial Regression : Captures non-linear relationships in the dataset.
*   Decision Tree Regression:Helps in identifying patterns that linear models may miss.
*   Random Forest Regression:An ensemble method that improves accuracy by combining multiple decision trees.

These models will be evaluated based on performance metrics such as Mean Absolute Error (MAE) and R² Score.

```python
# Compute correlation coefficients for all numerical variables
correlation_matrix = df_encoded.corr()
print(correlation_matrix)
```

```python
# Compute variance for numerical columns
variance = df_encoded.var()
print("\nVariance:\n", variance)

# Compute standard deviation
std_dev = df_encoded.std()
print("\nStandard Deviation:\n", std_dev)

# Compute covariance matrix
cov_matrix = df_encoded.cov()
print("\nCovariance Matrix:\n", cov_matrix)
```

```python
# Compute Coefficient of Variation (CV) = Standard Deviation / Mean
cv = (std_dev / df_encoded.mean()) * 100
print("\nCoefficient of Variation (CV) (%):\n", cv)
```
